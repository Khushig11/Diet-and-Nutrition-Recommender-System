```python
# Dropping unnecessary columns
data_cleaned = data.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'])

data_cleaned.head()
```

```python
# Selecting relevant features for recommendation
features = data_cleaned[['Caloric Value', 'Fat', 'Carbohydrates', 'Protein', 'Sugars', 'Dietary Fiber']]

# Normalize the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
from sklearn.neighbors import NearestNeighbors

# Building the Nearest Neighbors model
model = NearestNeighbors(n_neighbors=5, algorithm='ball_tree')
model.fit(features_scaled)

# Function to get recommendations
def recommend_diet(user_input):
    # Normalize user input
    user_input_scaled = scaler.transform([user_input])
    
    # Find the nearest neighbors
    distances, indices = model.kneighbors(user_input_scaled)
    
    # Get the recommended foods
    recommended_foods = data_cleaned.iloc[indices[0]]
    
    return recommended_foods

# Example user input [Calories, Fat, Carbohydrates, Protein, Sugars, Dietary Fiber]
user_input = [250, 10, 30, 15, 5, 10]

# Get recommendations
recommendations = recommend_diet(user_input)
print(recommendations)
```
